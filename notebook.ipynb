# Scraping Data

```python
# Library
import yfinance as yf
import pandas as pd
```

## Gold

```python
# fetch gold price data
df_gold = yf.download(
    "GC=F",
    period="max",
    interval="1d"
)

df_gold.head()
```

```python
df_gold.info()
```

```python
df_gold.to_csv("data/raw/gold.csv", index=False)
```

# Standarization data

## Rename data

```python
# Gold
df_gold.columns = df_gold.columns.get_level_values(0)

df_gold = df_gold.rename(columns={
    "Close": "gold_close",
    "High": "gold_high",
    "Low": "gold_low",
    "Open": "gold_open",
    "Volume": "gold_volume"
})

# IDR
df_idr.columns = df_idr.columns.get_level_values(0)

df_idr = df_idr.rename(columns={
    "Close": "idr_close",
    "High": "idr_high",
    "Low": "idr_low",
    "Open": "idr_open",
    "Volume": "idr_volume"
})

# Oil
df_oil.columns = df_oil.columns.get_level_values(0)

df_oil = df_oil.rename(columns={
    "Close": "oil_close",
    "High": "oil_high",
    "Low": "oil_low",
    "Open": "oil_open",
    "Volume": "oil_volume"
})

# DXY
df_dxy.columns = df_dxy.columns.get_level_values(0)

df_dxy = df_dxy.rename(columns={
    "Close": "dxy_close",
    "High": "dxy_high",
    "Low": "dxy_low",
    "Open": "dxy_open",
    "Volume": "dxy_volume"
})
```

```python
# Reset index
df_gold = df_gold.reset_index()
df_idr = df_idr.reset_index()
df_oil = df_oil.reset_index()
df_dxy = df_dxy.reset_index()
```
